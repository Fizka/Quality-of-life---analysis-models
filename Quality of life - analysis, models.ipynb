```python
from sklearn.decomposition import PCA

import numpy as np

boundary_x = list(range(0, 16))
boundary_y = [0.95 for _ in boundary_x]

pca = PCA().fit(X_scaled)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.scatter(boundary_x, boundary_y, marker='_', c='red')
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

```python
from sklearn.model_selection import train_test_split

# Podział na trenujące i testujące
training_data, testing_data, training_target, testing_target = train_test_split(df, y1, test_size=0.3, stratify=y1)
```

```python
from sklearn.neural_network import MLPClassifier

# Regresja logistyczna
from sklearn.linear_model import LogisticRegression

# Utworzenie sieci neuronowej
neural_network = MLPClassifier(hidden_layer_sizes=(100,50,20),activation ='identity',solver='adam',random_state=1)

# Standaryzacja i nauka modelu
neural_network.fit(preprocessing.StandardScaler().fit_transform(training_data), training_target)
```

```python
# Metryki poprawności
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score

# Seaborn is a library for making statistical graphics in Python
import seaborn as sns


# Utworzenie macierzy konfuzji
conf_matrix_neural_network = confusion_matrix(testing_target, \
                                              neural_network.predict(testing_data))
# Wizualizacja wyników
print("Confusion_matrix:")
print(conf_matrix_neural_network)
sns.heatmap(conf_matrix_neural_network)

# Wyświetlenie wyniku trafności modelu
acc = accuracy_score(testing_target, neural_network.predict(testing_data))
print("Neural network model accuracy is {0:0.2f}".format(acc))

# wyświetlenie f1-score
f1 = f1_score(testing_target, neural_network.predict(testing_data), average='micro')
print("F1 score is: {0:0.2f}".format(f1))
```

```python
from sklearn.tree import DecisionTreeClassifier

my_tree = DecisionTreeClassifier(criterion='entropy', max_depth=11);
my_tree.fit(training_data, training_target)
```

```python
# Utworzenie macierzy konfuzji
conf_matrix_neural_network = confusion_matrix(testing_target, \
                                              my_tree.predict(testing_data))
# Wizualizacja wyników
print("Confusion_matrix:")
print(conf_matrix_neural_network)
sns.heatmap(conf_matrix_neural_network)

# Wyświetlenie wyniku trafności modelu
acc = accuracy_score(testing_target, my_tree.predict(testing_data))
print("Neural network model accuracy is {0:0.2f}".format(acc))

# wyświetlenie f1-score
f1 = f1_score(testing_target, my_tree.predict(testing_data), average='micro')
print("F1 score is: {0:0.2f}".format(f1))
```

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=4)
for train_index, test_index in skf.split(df, y1):
    training_data1, testing_data1 = df.loc[train_index,:], df.loc[test_index,:]
    np.array(y1)[test_index]
    training_target1, testing_target1 = np.array(y1)[train_index], np.array(y1)[test_index]
```

```python
# Regresja logistyczna z walidacją krzyżową

neural_network = MLPClassifier(hidden_layer_sizes=(100,50,20),activation ='identity',solver='adam',random_state=1)
neural_network.fit(preprocessing.StandardScaler().fit_transform(training_data1), training_target1)
```

```python
conf_matrix_neural_network = confusion_matrix(testing_target1, neural_network.predict(testing_data1))
print("Confusion_matrix:")
print(conf_matrix_neural_network)
sns.heatmap(conf_matrix_neural_network)

acc = accuracy_score(testing_target1, neural_network.predict(testing_data1))
print("Neural network model accuracy is {0:0.2f}".format(acc))

f1 = f1_score(testing_target1, neural_network.predict(testing_data1), average='micro')
print("F1 score is: {0:0.2f}".format(f1))
```

```python
# drzewo decyzyjne z walidacją krzyżową

tree = DecisionTreeClassifier(criterion='entropy', max_depth=11);
tree.fit(training_data1, training_target1)
```
